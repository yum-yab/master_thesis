```julia
ps_data_monthly_path = joinpath(data_base_path, "ps_data_monthly")
psl_data_monthly_path = joinpath(data_base_path, "psl_data_monthly")
pr_data_monthly_path = joinpath(data_base_path, "preciptation_data_monthly")
ivt_data_monthly_path = joinpath(data_base_path, "ivt_fields_monthly")
eof_data_path = joinpath(data_base_path, "eof_data_v4")
```

```julia
function load_correlation_eof_piControl_data(scenario, eof_var, data_var; scopes="50seasons")

    mapping_var_path = Dict(
        ["ivt" => ivt_data_monthly_path, "pr" => pr_data_monthly_path, "psl" => psl_data_monthly_path]
    )

    return convert(Vector{Array{Float64,4}}, load("correlation_data/$(eof_var)_eof_$(data_var)_data_correlation_$(scenario).jld2")["data"]), load_eof_ensemble_common_scaling(eof_data_path, scopes, scenario, eof_var; scale_eofs=spatmodescale, align_with_first=2:5), build_ensemble_data_common_scaling(mapping_var_path[data_var], "piControl"; file_range_selection=1:13, data_field_id=data_var, member_range=1:1, filterfun=winter_timelimit)[1]
end
```

```julia
fig = Figure()

scenario = "ssp585"

eof_var = "psl"

data_var = "ivt"

correlation_data, eof_data, piControl = load_correlation_eof_piControl_data(scenario, eof_var, data_var)

eof_data_correlation_maps!(
    fig,
    eof_data,
    piControl,
    correlation_data;
    hexbin_colormap=:algae,
    colormap=:vik100,
    coastline_color=:black,
    shading=Makie.automatic,
)

fig
```
