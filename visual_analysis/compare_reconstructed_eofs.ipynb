```julia
function filter_winter_season(time_element)
    winter_months = [12, 1, 2, 3]
    for wm in winter_months
        if month(time_element) == wm
            return true
        end
    end
    return false
end
```

```julia
function get_comparison_of_mode(sign, comp_val, mode)

    function find_indices(temporal_data)

        result = Int[]

        standarized_data = standardize(ZScoreTransform, temporal_data[:, mode])
        # standarized_data = standardize(UnitRangeTransform, temporal_data[:, mode]; unit=false)


        println("Extrema: $(extrema(standarized_data))")


        for (i, val) in enumerate(standarized_data)


            if sign(val, comp_val)
                push!(result, i)
            end
        end
        return result
    end

    return find_indices
        
end
```

```julia
using GLMakie

fig = Figure()

lonmin, lonmax = extrema(ps_ssp585_full_timeline.lons)

latmin, latmax = extrema(ps_ssp585_full_timeline.lats)

ax1 = local_geoaxis_creation!(
    fig,
    (lonmin, lonmax),
    (latmin, latmax);
    figure_row=1,
    figure_col=1,
    title="EOF",
    dest_projection="+proj=merc",
)

ax2 = local_geoaxis_creation!(
    fig,
    (lonmin, lonmax),
    (latmin, latmax);
    figure_row=1,
    figure_col=3,
    title="DataMean",
    dest_projection="+proj=merc",
)

scope_index = 160

member_index = 1

my_scope = eof_ssp585_50seasons_results.scopes[scope_index]

eof_res = scale_eof_result(eof_ssp585_50seasons_results.ps_ensemble_eofs[get_member_id_string(member_index)][scope_index])

data_mean = dropdims(mean(ps_ssp585_full_timeline.members[member_index].data[:, :, my_scope], dims=3), dims=3) ./ 100

s_eof = surface!(ax1, lonmin .. lonmax, latmin .. latmax, eof_res.spatial_modes[:, :, 1] / 100, colormap = :vik100, shading=NoShading)

Colorbar(fig[1, 2], s_eof)

s_mean = surface!(ax2, lonmin .. lonmax, latmin .. latmax, data_mean / 100, colormap = :vik100, shading=NoShading)

Colorbar(fig[1, 4], s_mean)


DataInspector(fig)
```
