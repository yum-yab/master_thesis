```julia
using LinearAlgebra
```

```julia
data = rand(100, 600)
```

```julia
left_singular_vectors, singular_vals, right_singular_vectors = svd(data)
```

```julia
flips = [-1 ; ones(99)]
```

```julia
singular_vals
```

```julia
function reconstruct(left_singular_vectors, singular_vals, right_singular_vectors)
    return left_singular_vectors * Diagonal(singular_vals) * right_singular_vectors'
end
```

```julia
reconstructed = reconstruct(left_singular_vectors, singular_vals, right_singular_vectors)
```

```julia
diffs = data .- reconstructed

extrema(diffs)
```

```julia
right_singular_vectors
```

```julia
# now flip one axis

flipped_rights = right_singular_vectors .* reshape(flips, 1, :)
```

```julia
reconstructed_one_flipped = reconstruct(left_singular_vectors, singular_vals, flipped_rights)
```

```julia
reconstructed_both_flipped = reconstruct(left_singular_vectors .* reshape(flips, 1, :), singular_vals, flipped_rights)
```

```julia
maximum(data .- reconstructed_both_flipped)
```

```julia
maximum(data .- reconstructed_one_flipped)
```
