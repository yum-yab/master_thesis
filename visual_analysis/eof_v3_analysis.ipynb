```julia
function filter_winter_season(time_element)
    winter_months = [12, 1, 2, 3]
    for wm in winter_months
        if month(time_element) == wm
            return true
        end
    end
    return false
end
```

```julia
function winter_timelimit(timeelement)
    return filter_winter_season(timeelement) && timeelement < DateTime(2100, 12, 31)
end
```

```julia
get_variants() do scenario, scopes
    
    data_dict = Dict(variable => load_eof_ensemble(eof_data_path, scopes, scenario, variable; scale_eofs=spatmodescale, align_with_first=2:5) for variable in ["ivt", "psl", "pr"])

    pairings = [1 => 1, 1 => 2, 2 => 2, 1 => 3, 2 => 3, 3 => 3]

    for vartp in [("ivt", "psl"), ("pr", "ivt"), ("psl", "pr")]

        var1, var2 = vartp
        eof_data1 = data_dict[var1]

        eof_data2 = data_dict[var2]

        (_, scaling1, colormap1) = get_correct_var_display(var1)

        (_, scaling2, colormap2) = get_correct_var_display(var2)

        for (cor_type, use_crosscor) in [("crosscorrelation", true), ("correlation", false)]

            data_path = "ma_imgs/correlation_figures/$scenario/$scopes/$cor_type/$var1/"

            mkpath(data_path)

            for pair in pairings
                boxplot_fig = generate_correlation_boxplot(
                    eof_data1,
                    eof_data2;
                    compare_modes = pair, 
                    data_name = "Mode $(pair[1]) of Monthly $(uppercase(var1)) and Mode $(pair[1]) of Monthly $(uppercase(var1)) $(uppercase(scenario)) $(scopes)", 
                    size = (2100, 1000),
                    value_mode=:normal,
                    transformation=nothing,
                    legend_position = :rb,
                    show_first_pattern=true,
                    scalings=(scaling1, scaling2),
                    colormaps=(colormap1, colormap2),
                    use_crosscor=use_crosscor
                )

                file_path = joinpath(data_path, "$(cor_type)_boxplot_$(var1)_$(var2)_modes$(pair[1])$(pair[2])_$(scenario)_$(scopes).png")

                save(file_path, boxplot_fig)

            end


        end
      
    end
    
end
```

```julia
function get_correlation_map(variable_data, temporal_pattern)

    x, y = size(variable_data)

    println("$x, $y")

    result = ones(x, y)

    for i in 1:x
        for j in 1:y
            result[i, j] = cor(variable_data[i, j, :], temporal_pattern)
        end
    end
    
    return result
end
```
