```julia
function get_contour_lines_eof_ensemble(lons, lats, eof_ensemble, time_index; mode_selection = 1, line_selection = :)

    xvals = Float64[]
    yvals = Float64[]
    for (member, eof_data) in eof_ensemble.ivt_ensemble_eofs
        frame = eof_data[time_index].spatial_modes[:, :, mode_selection]
        clines = Contour.lines(levels(contours(ps_ssp585_monthly.lons, ps_ssp585_monthly.lats, frame, [0]))[1])[line_selection]
        new_x, new_y = unzip([(x, y) for l in clines for (x, y) in l.vertices  if !ismissing(x) && !ismissing(y)]::Vector{Tuple{Float64, Float64}})

        append!(xvals, new_x)
        append!(yvals, new_y)
    end
    return xvals, yvals
end
```

```julia
function get_contour_lines_per_member(lons, lats, eof_ensemble, time_index; mode_selection = 1, line_selection = :)

    results = Dict{String, Vector{Tuple{Vector{Float64}, Vector{Float64}}}}()

    for (member, eof_data) in eof_ensemble.ivt_ensemble_eofs
        frame = eof_data[time_index].spatial_modes[:, :, mode_selection]
        clines = Contour.lines(levels(contours(ps_ssp585_monthly.lons, ps_ssp585_monthly.lats, frame, [0]))[1])[line_selection]
        res = Tuple{Vector{Float64}, Vector{Float64}}[]
        for l in clines
            new_x, new_y = unzip([(x, y) for (x, y) in l.vertices  if !ismissing(x) && !ismissing(y)]::Vector{Tuple{Float64, Float64}})
            push!(res, (new_x, new_y))
        end
        results[member] = res
    end
    return results
end
```

```julia
function sample_along_line(line, dx)
    n = length(line)
    sampled_points = []
    
    # Calculate total length of the line
    total_length = 0.0
    for i in 1:(n - 1)
        total_length += norm([line[i+1][1] - line[i][1], line[i+1][2] - line[i][2]])
    end
    
    # Interpolate and sample along the line
    distance_covered = 0.0
    for i in 1:(n - 1)
        (x1, y1) = line[i]
        (x2, y2) = line[i + 1]
        segment_length = norm([x2 - x1, y2 - y1])
        
        while distance_covered < segment_length
            t = distance_covered / segment_length
            x_sample = (1 - t) * x1 + t * x2
            y_sample = (1 - t) * y1 + t * y2
            push!(sampled_points, (x_sample, y_sample))
            distance_covered += dx
        end
        
        # Adjust the distance covered for the next segment
        distance_covered -= segment_length
    end
    
    # Add the last point of the line
    push!(sampled_points, line[end])
    
    return sampled_points
end
```

```julia
function transparent_colormap(n::Int)
    colors = [RGBA(1.0, 1.0, 1.0, 0.0) for i in 1:n]
    for i in 1:n
        t = (i - 1) / (n - 1)
        colors[i] = RGBA(1.0, 0.0, 0.0, t)  # Transition from white (transparent) to red (opaque)
    end
    return colors
end

# Create the colormap with 256 steps
custom_colormap = transparent_colormap(50)
```
